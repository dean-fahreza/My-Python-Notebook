# Dataset Preparation

```python
# membuat salinan dataframe
df = data.copy(deep = True)
```

```python
# mengubah/convert nilai "?" nilai ke bentuk Na / NaN untuk diproses lebih lanjut
for col in data.columns:
    df[[col]] = data[[col]].replace('?', np.NaN)
```

```python
# seleksi kolom fitur atau features dari tabel dataset
null_data = df.iloc[:, :-1]
```

```python
# menemukan nilai null untuk semua attribut dan jumlah total nilai null
null_data.isnull().sum()
```

```python
# menghapus semua baris tabel dataset yang mempunyai nilai "null"
df = df.dropna()
```

```python
# pilih kolom fitur atau features dari tabel dataset
null_data = df.iloc[:, :-1]
```

```python
# cek ulang nilai null
null_data.isnull().sum()
```

**StandardScaler** *adalah class dari sklearn untuk melakukan normalisasi data agar data yang digunakan tidak memiliki penyimpangan yang besar.*

```python
# mengimport library standard scaler 
from sklearn.preprocessing import StandardScaler
```

```python
# membuat dataframe dengan tipe data int64
colname = []
for col in data.columns:
    if df[col].dtype == 'int64':
        colname.append(col)
```

```python
# membuat salinan dataset untuk keperluan persiapan data / data preparation
df_copy = df.copy(deep = True)
df_fe = df.copy()
```

```python
# membuat kerangka data untuk fitur kategoris / categorical features
df_fe.drop('income',axis='columns', inplace=True)
df_fe.drop(colname,axis='columns', inplace=True)
```

```python
# membuat dataframe untuk kelas target / target class
df_cl = df.copy()
df_cl.drop(df_copy.iloc[:,:-1],axis='columns', inplace=True)
```

```python
# membuat objek scaler / scaler object
std_scaler = StandardScaler()
std_scaler
```

```python
# normalisasikan atribut numerik dan tetapkan ke dalam dataframe baru
df_norm = pd.DataFrame(std_scaler.fit_transform(df_copy[colname]), columns=colname)
```

```python
# mengimport library Ordinal Encoder 
from sklearn.preprocessing import OrdinalEncoder
ord_enc = OrdinalEncoder()
```

```python
# enconde fitur kategoris/categorical features menjadi fitur numerik/numerical features
for col in df_fe.columns[:]:
    if df_fe[col].dtype == 'object':
        df_fe[col] = ord_enc.fit_transform(df_fe[[col]])
```

```python
# enconde label kategorikal/categorical label menjadi label biner/binary label 
df_cl['income'] = np.where(df_cl['income'].str.contains('>50k'), 0, 1)
```

```python
# masukkan kolom id ke datasets yang berbeda
df_norm.insert(0, 'id', range(0, 0 + len(df_norm)))
df_fe.insert(0, 'id', range(0, 0 + len(df_fe)))
df_cl.insert(0, 'id', range(0, 0 + len(df_cl)))
```

```python
# melihat shape databases yang telah di proses
print(df_norm.shape)
print(df_fe.shape)
print(df_cl.shape)
```

```python
# menggabungkan semua datasets
df_feature = pd.merge(df_norm, df_fe, on=['id'])
df_final = pd.merge(df_feature, df_cl, on=['id'])
```

```python
# hapus kolom 'id' dari gabungan datasets
df_final.drop('id', axis='columns', inplace=True)
```

```python
# melihat lima data dari gabungan datasets
df_final.head(5)
```

# Proses Modelling dengan KKN

```python
# mengimport packages
from sklearn.model_selection import KFold
```

```python
X = df_final.iloc[:,:].to_numpy()
y = df_final.iloc[:,-1:].to_numpy()
kf = KFold(n_splits=5)
kf.get_n_splits(X)

print(kf)

for train_index, test_index in kf.split(X):
    print("TRAIN:", train_index, "TEST:", test_index)
    x_train, x_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

```python
print('-------- x axis test ----------')
print(x_test)
print('-------- x axis train ---------')
print(x_train)
print('-------- y axis test ----------')
print(y_test)
print('-------- y axis train ---------')
print(y_train)
print('*******************************')
```

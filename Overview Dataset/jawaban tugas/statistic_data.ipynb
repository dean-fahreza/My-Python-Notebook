# 3. Histogram

```python
## Buatlah histogram dari kolom strength
x = plt.hist(data['strength'], bins=30, fc=(0.2, 1, 0.1, 0.8))
plt.xlabel = ("Nilai Data dari Kolom Strength")
plt.ylabel = ("Frequency")
plt.title = ("Histogram Frequency")
plt.show()
```

# 4. Box Plot

```python
## Buatlah histogram dari kolom strength
strength = pd.melt(data, id_vars=None, value_vars=['strength'])
plt.figure(figsize=(20,15))
sns.boxplot(x = 'variable', y = 'value', data = strength)
plt.show
```

# 5. Summary Statistics

```python
## Hitung karakteristik data dari masing-masing kolom dengan menggunakan perintah describe.
# dapat dilihat data-data tersebut menggunakan method data.head() dan data.columns

# cement
print("--- Cement ---")
print("Standard Deviation (std) : ", data['cement'].std())
print("Variance (var) : ", data['cement'].var())
print("Mean (mean) : ", data['cement'].mean())
print("Median (median) : ", data['cement'].median())

# slag
print("\n--- Slag ---")
print("Standard Deviation (std) : ", data['slag'].std())
print("Variance (var) : ", data['slag'].var())
print("Mean (mean) : ", data['slag'].mean())
print("Median (median) : ", data['slag'].median())

# ash
print("\n--- Ash ---")
print("Standard Deviation (std) : ", data['ash'].std())
print("Variance (var) : ", data['ash'].std())
print("Mean (mnean) : ", data['ash'].mean())
print("Median (median) : ", data['ash'].median())

# water
print("\n--- Water ---")
print("Standard Deviation (std) : ", data['water'].std())
print("Variance (var) : ", data['water'].var())
print("Mean (mean) : ", data['water'].mean())
print("Median (median) : ", data['water'].median())

# superplastic
print("\n--- Super Plastic ---")
print("Standard Deviation (std) : ", data['superplastic'].std())
print("Variance (var) : ", data['superplastic'].var())
print("Mean (mean) : ", data['superplastic'].mean())
print("Median (median) : ", data['superplastic'].median())

# coarseagg
print("\n--- Coarse Agg ---")
print("Standard Deviation (std) : ", data['coarseagg'].std())
print("Variance (var) : ", data['coarseagg'].var())
print("Mean (mean) : ", data['coarseagg'].mean())
print("Median (median) : ", data['coarseagg'].median())

# fineagg
print("\n--- Fine Agg ---")
print("Standard Deviation (std) : ", data['fineagg'].std())
print("Variance (var) : ", data['fineagg'].var())
print("Mean (mean) : ", data['fineagg'].mean())
print("Median (median) : ", data['fineagg'].median())

# age
print("\n--- Age ---")
print("Standard Deviation (std) : " , data['age'].std())
print("Variance (var) : ", data['age'].var())
print("Mean (mean) : ", data['age'].mean())
print("Median (median) : ", data['age'].median())

# strength
print("\n--- Strength ---")
print("Standard Deviation (std) : ", data['strength'].std())
print("Variance (var) : ", data['strength'].var())
print("Mean (mean) : ", data['strength'].mean())
print("Median (median) : ", data['strength'].median())
```

# 6. Relationship Between Variables

```python
# Tampilkan hubungan antara data kolom 'age' dan 'strength'
plt.figure(figsize=(20,15))
sns.jointplot(data.age, data.strength, kind="reg")
plt.show()
```

```python
# Tampilkan hubungan antara data kolom 'cement', 'slag','age' dan 'strength'
sns.set(style = 'white')
df = data.loc[:,['cement', 'slag','age','strength']]
g = sns.PairGrid(df, diag_sharey = False,)
g.map_lower(sns.kdeplot, cmap="Greens_d")
g.map_upper(plt.scatter)
g.map_diag(sns.kdeplot, lw = 4)
plt.show
```

# 7. Correlation

```python
## Buatlah diagram heatmap dari setiap kolom yang ada dengan library seaborn
f,ax=plt.subplots(figsize = (18,18))
sns.heatmap(data.corr(),annot= True,linewidths=0.5,fmt = ".1f",ax=ax)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.title('Correlation Map')
plt.savefig('graph.png')
plt.show()
```

# 8. Covariance

```python
## Bandingkan nilai covariance dari data age, strength dan strength dan cement
np.cov(data.age, data.strength)
print("Covariance diantara Age dan Strenght : ", data.age.cov(data.strength))
print("Covariance diantara Strength dan Cement : ", data.strength.cov(data.cement))
```

# 9. Pearson Correlation

```python
## Hitung nilai spearsman rank dari kolom data age dan strength
ranked_data = data.rank()
spearman_corr = ranked_data.loc[:,["age","strength"]].corr(method= "pearson")
print("Spearman's correlation: ")
print(spearman_corr)
```

# 10. Hypothesis Testing

```python
## Lakukan hubungan hipotesis data antara kolom age dan strength
statistic, p_value = stats.ttest_rel(data.age,data.strength)
print('p-value adalah: ',p_value)
```
